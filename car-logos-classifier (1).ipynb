```python
import fastai
from fastai import *
from fastai.vision.all import *
from fastai.vision.widgets import *
```

```python
import urllib.request

MODEL_URL = "https://drive.google.com/uc?export=download&id=1Yu9DUjjyn4POBRvexswHKVIgi-o8aMDI"
x = urllib.request.urlretrieve(MODEL_URL, "export.pkl")

learner = load_learner("export.pkl", cpu = True)
```

```python
btn_upload = widgets.FileUpload()
out_pl = widgets.Output()
lbl_pred = widgets.Label()
```

```python
def on_data_change(change):
    lbl_pred.value = ''
    img = PILImage.create(btn_upload.data[-1])
    out_pl.clear_output()
    with out_pl: display(img.to_thumb(128,128))
    pred,pred_idx,probs = learn_inf.predict(img)
    lbl_pred.value = f'Prediction: {pred}; We think it is a {pred} car with a probability of {probs[pred_idx]:.04f}'
```

```python
btn_upload.observe(on_data_change, names=['data'])
```

```python
display(VBox([widgets.Label('What car is that?'), btn_upload, out_pl, lbl_pred]))
```
